```python
clevr_instances_file = osp.join("/home/arjunakula/Dropbox/My_UCLA_docs_from_2016_sept/PhD_Research/after_summer_2019/EMNLP2020_Mila/clevr_ref+/iep-ref/data/instances_new.json")
instances = json.load(open(clevr_instances_file, 'r'))
```

```python
refs = pickle.load(open('/home/arjunakula/Dropbox/My_UCLA_docs_from_2016_sept/PhD_Research/after_summer_2019/EMNLP2020_Mila/clevr_ref+/iep-ref/data/refs(clevr).p', 'r'))
```

```python
#build a map from annotation id to annottaion
annotation_map = {}
for ann in instances['annotations']:
    annotation_map[ann['id']] = ann
```

```python
annotation_map
```

```python
all_detections = []
gt_detections = []

start_det_id = 0
for ann in instances['annotations']:
    new_det = {}
    new_det['box']=ann['bbox']
    new_det['h5_id']=start_det_id
    new_det['category_name']=""
    new_det['image_id']=ann['image_id']
    new_det['image_filename']=ann['image_filename']
    new_det['score']=1
    new_det['category_id']=1
    new_det['det_id']=start_det_id
    start_det_id = start_det_id + 1
    all_detections.append(new_det)
    
for ref in refs:
    new_det = {}
    new_det['refBox']=annotation_map[ref['ann_id']]['bbox']
    new_det['image_id']=ref['image_id']
    new_det['ref_id']=ref['ref_id']
    gt_detections.append(new_det)
```
